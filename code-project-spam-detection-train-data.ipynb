```python
#Lowercase
df['Heading']=df['Heading'].str.lower()
df['MainText']=df['MainText'].str.lower()
df.head()
```

```python
#Punctuation removal
punct='!"$%&\'()*,-./:;<=>?@[\\]^_{|}~`'

def remove_punct(text):
    return text.translate(str.maketrans('','',punct))
    
```

```python
#Replacing NA value 
df['Heading']=df['Heading'].fillna('')
df['MainText']=df['MainText'].fillna('')
df['PrimarySubject']=df['PrimarySubject'].fillna('')
df['SecondarySubject']=df['SecondarySubject'].fillna('')
df['TertiarySubject']=df['TertiarySubject'].fillna('')
df['OtherSubject']=df['OtherSubject'].fillna('')
```

```python
df['Heading']=df['Heading'].apply(lambda text:remove_punct(text))
df['MainText']=df['MainText'].apply(lambda text:remove_punct(text))
```

```python
df['PrimarySubject']=df['PrimarySubject'].apply(lambda text:remove_punct(text))
df['SecondarySubject']=df['SecondarySubject'].apply(lambda text:remove_punct(text))
df['TertiarySubject']=df['TertiarySubject'].apply(lambda text:remove_punct(text))
df['OtherSubject']=df['OtherSubject'].apply(lambda text:remove_punct(text))
```

```python
df.head()
```

```python
#Removal of stopwords

from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
#newlist = [expression for item in iterable if condition == True]
#string.join(iterable)
#lambda args: expression

STOPWORDS=set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
df['Heading']=df['Heading'].apply(lambda text:remove_stopwords(text))
df['MainText']=df['MainText'].apply(lambda text:remove_stopwords(text))
```

```python
from collections import Counter
cnt=Counter()

for text in df['Heading'].values:
    for word in text.split():
     cnt[word] +=1;
    
    
cnt.most_common(10)   
```

```python
#Removal of frequent words
FREQWORDS=set([w for (w,wc) in cnt.most_common(1)])

def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df['Heading']=df['Heading'].apply(lambda text: remove_freqwords(text))
```

```python
from collections import Counter
cnt=Counter()

for text in df['MainText'].values:
    for word in text.split():
     cnt[word] +=1;
    
    
cnt.most_common(10)   
```

```python
FREQWORDS=set([w for (w,wc) in cnt.most_common(10)])

def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df['MainText']=df['MainText'].apply(lambda text: remove_freqwords(text))
```

```python
#Lemmatization is a bit more complex in that the computer can group together words that do not have the same stem, but still have the same inflected meaning

from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

def lemmatize_word(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['Heading']=df['Heading'].apply(lambda text: lemmatize_word(text))
```

```python
df.head()
```

```python
df['MainText']=df['MainText'].apply(lambda text: lemmatize_word(text))
```

```python
#Removal of URLs
#(?:xyz) Non-capturing group of characters
#\S Matches a non-whitespace character
#+  Matches 1 or more of the previous

def remove_urls(text):
    url_pattern=re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'',text)                       
                           
```

```python
df['Heading']=df['Heading'].apply(lambda text:remove_urls(text))
df['MainText']=df['MainText'].apply(lambda text:remove_urls(text))
```

```python
#Normalizing date time columns

df['Postyear']=(df['Postyear']-df['Postyear'].mean())/df['Postyear'].std()
df['Postmonth']=(df['Postmonth']-df['Postmonth'].mean())/df['Postmonth'].std()
df['Postday']=(df['Postday']-df['Postday'].mean())/df['Postday'].std()
df['Posthour']=(df['Posthour']-df['Posthour'].mean())/df['Posthour'].std()
df['Postminute']=(df['Postminute']-df['Postminute'].mean())/df['Postminute'].std()

df['Accountyear']=(df['Accountyear']-df['Accountyear'].mean())/df['Accountyear'].std()
df['Accountmonth']=(df['Accountmonth']-df['Accountmonth'].mean())/df['Accountmonth'].std()
df['Accountday']=(df['Accountday']-df['Accountday'].mean())/df['Accountday'].std()
df['Accounthour']=(df['Accounthour']-df['Accounthour'].mean())/df['Accounthour'].std()
df['Accountminute']=(df['Accountminute']-df['Accountminute'].mean())/df['Accountminute'].std()


df['Karma']=(df['Karma']-df['Karma'].mean())/df['Karma'].std()
df['NumAnswers']=(df['NumAnswers']-df['NumAnswers'].mean())/df['NumAnswers'].std()
```

```python
df
```

```python
#Concatenating the columns 

df_concat=df['Heading']+' '+df['MainText']+' '+df['PrimarySubject']+' '+df['SecondarySubject']+' '+df['TertiarySubject']+' '+df['OtherSubject']

df_concat
```

```python
#Vectorization  

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=TfidfVectorizer(lowercase=False,max_features=1000)
vectorizer.fit(df_concat)


```

```python
#Sparse matrix having train dataset tf-idf values
X_Tfidf = vectorizer.transform(df_concat)
```

```python
import scipy as sp


X_Tfidf 
```

```python
X_Tfidf.shape
```

```python
rows=X_Tfidf.shape[0]
```

```python
rows
```

```python
#Adding features of dataframe df

#scipy.sparse provides functions to deal with sparse data

#Feature addition to the sparse matrix

#scipy.sparse.hstack--->Stack sparse matrices horizontally

df_temp=df['Karma'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['NumAnswers'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Postyear'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Postday'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Posthour'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Postminute'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Accountyear'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Accountmonth'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Accountday'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Accounthour'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

df_temp=df['Accountminute'].to_numpy().reshape(rows,-1)
X_Tfidf=sp.sparse.hstack((X_Tfidf,df_temp))

```
